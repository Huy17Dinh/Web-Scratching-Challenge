```python
import requests
import os
import pandas as pd
import time
from bs4 import BeautifulSoup as bs
from splinter import Browser
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager
```

```python
#browser = initial broswer
excutable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **excutable_path, headless=False)
```

```python
news_url = 'https://redplanetscience.com/'
browser.visit(news_url)
```

```python
#BeatifulSoup object
news_html = browser.html
news_soup = bs(news_html, "html.parser")
print(news_soup)
```

```python
#extract titles and paragraphs
news_titles = news_soup.find("div", class_="content_title").text
news_para = news_soup.find("div", class_="article_teaser_body").text
print(f'Latest news: {news_titles}')
print(f'Description: {news_para}')
```

```python
hem_url = 'https://marshemispheres.com/'
browser.visit(hem_url)
hem_html = browser.html
hem_soup = bs(hem_html, 'html.parser')
results = hem_soup.find_all("div", class_='item')
```

```python
hem_image_url = []
for result in results:
    product_dict = {}
    titles = result.find('h3').text
    end_link = result.find('a')['href']
    img_link = "https://marshemispheres.com/" + end_link 
    browser.visit(img_link)
    html = browser.html
    soup = bs(html, 'html.parser')
    download = soup.find("div", class_="downloads")
    image_url = "https://marshemispheres.com/" + download.find("img")["src"]
    print(titles)
    print(image_url)
    product_dict['title'] = titles
    product_dict['image_url'] = image_url
    hem_image_url.append(product_dict)
```
